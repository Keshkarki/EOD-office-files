## pivot and pivot table

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
round(df.pivot_table(index='sex', columns='smoker', values='tip', aggfunc='mean', margins=True),2)
```

```python
round(df.pivot_table(index='sex', columns='smoker', values='tip', aggfunc='sum', margins=True),2)
```

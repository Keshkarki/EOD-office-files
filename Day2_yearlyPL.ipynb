## Yearly P&L 

```python
import pandas as pd
import numpy as np
```

## rolling avg and reading file

```python
rolling_avg = 20
df = pd.read_csv("C:\\keshav\\Nifty50_data\\Nifty50.csv", parse_dates=['Date'],index_col='Date').sort_index()
df = df.loc['2018-04-02':]
df
```

```python
df['Close_MA'] =df['Close'].rolling(rolling_avg).mean()  #rolling _avg variable given 20 above
```

```python
df.head(30)
```

```python
def get_signal(row):
    if row['Close'] > row['Close_MA']:
        return 'Buy'
    elif row['Close'] < row['Close_MA']:
        return 'Sell'
    else:
        return '-'

df['Signal'] = df.apply(get_signal, axis=1)
```

```python
# df['Signal'] = df.apply(lambda row: 'Buy' if (row['Close'] > row['Close_MA']) else ('Sell' if row['Close'] < row['Close_MA'] else '-'), axis=1)
```

```python
df.head(30)
```

```python
#still will buy or sell next day hence shifting one down to singnal
df['Signal'] = df['Signal'].shift(1)
```

```python
df.head(30)
```

```python
df['Signal2'] = df['Signal'].shift(1)
```

```python
df.head(50)
```

```python
# whenever signal changes track next day open value
def calc_change(row):
    if row['Signal'] != row['Signal2'] and row['Signal'] != '-' and row['Signal2'] is not None:
        return row['Open']
    else:
        return '-'

df['change'] = df.apply(calc_change, axis=1)
```

```python

# df['change'] = df.apply(lambda row: row['Open'] if row['Signal']!=row['Signal2'] and row['Signal']!='-' and  row['Signal2'] is not None else '-', axis=1)
df.head(30)
```

```python
# convering change column to floats so we can filter it out
df['change'] = df.apply(lambda x: float(x['change']) if x['change'] != '-' else 0, axis=1)
```

```python
df.head(30)
```

## filtered dataframe

```python
# filtering out rows based on change column having open values only
filtered_df = df[df['change']>0].copy()
filtered_df
```

```python
filtered_df['shifted_change'] = filtered_df['change'].shift(1)
```

```python
filtered_df
```

```python
filtered_df['P&L'] = filtered_df.apply(lambda row: (row['change'] - row['shifted_change']) if row['Signal'] == 'Sell' else row['shifted_change']-row['change'],axis=1)
```

```python
filtered_df
```

```python
#ignoring first row not having any position for that
filtered_df = filtered_df[1:]
```

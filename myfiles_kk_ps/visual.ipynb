```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df = pd.read_excel("C:\\Users\\Admin\\OneDrive\\Desktop\\visual.xlsx")

df.rename(columns={'Unnamed: 0':'date'},inplace=True)
df['NIFTY'] = round(df['NIFTY'],2)
df['Product'] = round(df['Product'],2)
df= df[['date','NIFTY','Product']]
df['date'] = df['date'].dt.year.astype(str)
df.set_index('date', inplace=True)
df
```

```python

fig, ax = plt.subplots(figsize=(12, 6))

# colors = ['#F5460E', '#5BDF0F']
 
# Plot the stacked bar chart
df.plot(kind='bar', stacked=True, ax=ax)

# Add values at the top of each segment
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=8, fmt='%.2f')

# Add labels and a title
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.axhline(color='k')

plt.show()
```

```python
# Increase the figure size

fig, ax = plt.subplots(figsize=(12, 6))

colors = ['#F5460E', '#5BDF0F']
 
# Plot the stacked bar chart
df.plot(kind='bar', stacked=True, ax=ax,color=colors)

# Add values at the top of each segment
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=8, fmt='%.2f')

# Add labels and a title
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.axhline(color='k')

plt.show()
```

```python
import numpy as np
import pandas as pd
import os

pd.set_option('display.max_columns',None)
pd.set_option('display.width', None)
# pd.set_option('display.max_rows',None)
```

##  nifty50 file reading

```python
niftySingle = pd.read_csv('niftySingle.csv', usecols=['Date', 'Close'], index_col='Date')
niftySingle.rename(columns={'Close' : 'nifty_price'},inplace=True)
niftySingle
```

## list of all companies

```python
list_of_companies = os.listdir('C:\\keshav\\Navneet_sir_all\\Nifty50_dashboard_levels\\Nifty50_Data')
list_of_companies
```

## merging 

```python
combined_df = pd.DataFrame()
for company  in list_of_companies:
    print(company)

    df = pd.read_csv(f'C:\\keshav\\Navneet_sir_all\\Nifty50_dashboard_levels\\Nifty50_Data\\{company}', usecols=['Date','Close'], index_col='Date')
    company_name = f"{company.split('.')[0]}_"

    df.rename(columns={'Close' : f"{company_name}price"}, inplace=True)
    # df.insert(0,'company',company_name)

    dff = pd.merge(niftySingle,df ,left_index=True , right_index=True)
    dff[f"{company_name}ratio"] = dff['nifty_price']/dff[f"{company_name}price"]
    dff[f"{company_name}mean"] = dff[f"{company_name}ratio"].mean()
    dff[f"{company_name}std"] = dff[f"{company_name}ratio"].std()
    combined_df = pd.concat([combined_df,dff], axis =1)
    
combined_df
```

```python
combined_df.columns
```

### removing all aother nifty_price columns

```python

# Get the column names
columns = combined_df.columns
# print(columns)

# Find the indices of columns with 'nifty_price' in their names
indices_to_drop = [i for i, col in enumerate(columns) if col == 'nifty_price']
print(indices_to_drop)

# Drop the columns except the first one
combined_df = combined_df.drop(columns=columns[indices_to_drop[1:]])
```

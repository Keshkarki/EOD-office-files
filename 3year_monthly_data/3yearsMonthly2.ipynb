## reading csv's

```python
df_22 = pd.read_csv("C:\\keshav\\3year_monthly_data\\fy21-22.csv")
df_22 = df_22.iloc[1:]
df_22.head()
```

```python
df_23 = pd.read_csv("C:\\keshav\\3year_monthly_data\\fy22-23.csv")
df_23.head()
```

## Third column  manupulation

```python
df_23.drop(columns=['ISIN','BuyQty','BuyRate','BuyVal','SellQty','SellRate','SellVal','LTPRate','LTPVal','IncomeTax'],inplace=True)
```

```python
column_names = ['Instrument', 'SoldQuantity', 'BuyAvg', 'BuyValue', 'SellAvg', 'SellValue', 'NetRealizedPnL', 'RealisedGLPerc', 'CloseQuantityAsOnLastDay', 'BuyAvgOpenPrice', 'SellAvgPrice', 'CMP', 'NetUnrealizedPnL', 'NetUnrealizedPnlPerc', 'TotalGL', 'TotalGLPerc', 'DayGL', 'DayGLPerc', 'AssetType']
```

```python
temp_df = pd.DataFrame(columns=column_names)
temp_df
```

```python
temp_df['Instrument'] = df_23['ScriptName'].values
temp_df['NetRealizedPnL'] = df_23['NetRealizedPnL'].values
temp_df.fillna(0,inplace=True)
```

```python
temp_df.head()
```

### applying regex 

```python
import re
temp_df['Date'] = pd.to_datetime(temp_df['Instrument'].apply(lambda x: re.search(r'\d{2}[A-Za-z]{3}\d{4}', x).group(0)))
temp_df.set_index('Date',inplace=True)
```

```python
temp_df.head(2)
```

### checking shape

```python
print(df_21.shape)
print(df_22.shape)
print(temp_df.shape)
```

## concatenating df1 and df2

```python
df = pd.concat([df_21,df_22], axis=0)
```

### regex extracting date

```python
import datetime as dt
df['Date'] = df['Instrument'].str.extract(r'(\d{2}[A-Za-z]{3}\d{4})')
df["Date"] = pd.to_datetime(df['Date'])
df.set_index('Date',inplace=True)
```

## concate these two df with temp_df

```python
merged_df = pd.concat([df,temp_df], axis=0).sort_index()
```

```python
merged_df.insert(0,"Year",merged_df.index.year.to_list())
merged_df.insert(1,"Month",merged_df.index.month.to_list())
```

## merged dataframe

```python
merged_df
```

### total sum

```python
merged_df['NetRealizedPnL'].sum()
```

### group by year and month

```python
round(merged_df.groupby(['Year','Month'])['NetRealizedPnL'].sum())
```

### grp_yearly

```python
grp_yearly = round(merged_df.groupby(['Year','Month'])['NetRealizedPnL'].sum()).reset_index()
grp_yearly
```

```python
# NetRealizedPnl in per
grp_yearly['NetRealizedPnLPerc'] = round(grp_yearly['NetRealizedPnL']/300000,2)
```

```python
grp_yearly
```

### group by yearly with percentage

```python
grp_yearly = grp_yearly[['Year','Month','NetRealizedPnLPerc']]
grp_yearly
```

## Nifty data

### reading nifty

```python
nifty = pd.read_csv('Nifty50.csv')
nifty
```

```python
nifty['Date'] = pd.to_datetime(nifty['Date'])
nifty.set_index('Date',inplace=True)
```

```python
nifty = nifty.loc[merged_df.index[0]:merged_df.index[-1]]
```

```python
nifty['symbol'] = 'Index'
```

```python
nifty
```

## monthly data 

```python
# Resample daily data to monthly data
monthly_data = nifty.resample('BM').last()

# Calculate monthly returns
monthly_data['monthly_return'] = ((monthly_data['Adj Close'] - monthly_data['Adj Close'].shift(1)) / monthly_data['Adj Close'].shift(1))*100

# Drop the first row as it will have NaN value
monthly_data = monthly_data.dropna()

# Print the monthly returns
print(monthly_data['monthly_return'])
```

```python
monthly_data
```

```python
monthly_data.insert(0,'Year',monthly_data.index.year)
```

```python
monthly_data.insert(1,'Month',monthly_data.index.month)
```

```python
monthly_data
```

```python
monthly_data = round(monthly_data.groupby(['Year','Month'])['monthly_return'].sum().reset_index())
monthly_data
```

## combining yearly data with monthly index data

```python
stockVSindex = pd.merge(grp_yearly, monthly_data, on=["Year", "Month"], how="left")
stockVSindex
```

```python
stockVSindex.rename(columns={
    'NetRealizedPnLPerc': 'stock_NetRealizedPnLPerc',
    'monthly_return': 'index_monthly_return'
},inplace=True)
```

```python
stockVSindex
```

```python
stockVSindex.to_csv('proft_stock_vs_index_grp_monthly.csv')
```

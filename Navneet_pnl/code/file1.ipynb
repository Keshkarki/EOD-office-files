## importing libraries

```python
import numpy as np
import pandas as pd
```

## read os dir

```python
import os
files_list = os.listdir("C:\\keshav\\Navneet_pnl\\all_files")

len(files_list)
```

```python
final_df = pd.DataFrame()
processed_files = []

for file in files_list:
    try:
        if file in processed_files:
            print(f"Skipping duplicate file: {file}")
            continue

        df = pd.read_excel(f"C:\\keshav\\Navneet_pnl\\all_files\\{file}", header=None)

        date = df.iloc[21:22, 3:4][3].values[0]
        NetRealizedPnL = df.iloc[25:26, :][8].values[0]
        NetUnrealizedPnlPerc = df.iloc[25:26, :][14].values[0]
        TotalGL = df.iloc[25:26, :][16].values[0]

        empty_df = pd.DataFrame(columns=['date', 'NetRealizedPnL', 'NetUnrealizedPnlPerc', 'TotalGL'])

        empty_df.loc[0, 'date'] = date
        empty_df.loc[0, 'NetRealizedPnL'] = NetRealizedPnL
        empty_df.loc[0, 'NetUnrealizedPnlPerc'] = NetUnrealizedPnlPerc
        empty_df.loc[0, 'TotalGL'] = TotalGL

        final_df = pd.concat([final_df, empty_df])
        processed_files.append(file)

    except FileNotFoundError:
        print('File Not Found')
    except Exception as e:
        print(f'Error occurred while processing {file}: {e}')
```

##  beautifying


```python
final_df['date']  = final_df['date'].astype(str)
final_df.insert(0,'from_date',final_df['date'].str.split(' to ').str.get(0))
final_df.insert(1,'to_date',final_df['date'].str.split(' to ').str.get(1))
final_df.drop('date',axis=1,inplace=True)
```

## date sort

```python
final_df['from_date'] = pd.to_datetime(final_df['from_date'], format='%d-%b-%Y')
final_df['to_date'] = pd.to_datetime(final_df['to_date'], format='%d-%b-%Y')

final_df = final_df.sort_values(by='from_date', ascending=True)
```

## Exporting to csv

```python
final_df.to_csv('PnLfile_19added.csv',index=False)
```

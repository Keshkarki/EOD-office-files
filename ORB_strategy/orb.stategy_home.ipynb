```python
"""PART 1- CLEANUP!!"""
df = pd.read_csv("C:\\keshav\\Intraday_1_minute\\2022\\2022 APR BNF.txt",index_col=False,header=None)
df = df.iloc[:,:7]

df.rename(columns={

            0:'index',
            1 : 'date',
            2: 'time',
            3: 'open',
            4:'high',
            5:'low',
            6:'close'
},inplace=True)
```

```python
#pre setup str for concatenation further
df['date'] = df['date'].astype('str')
datetime = pd.to_datetime(df['date'] + ' ' + df['time'], format='%Y%m%d %H:%M')
df.insert(1,'datetime',datetime) 
df['date'] = pd.to_datetime(df['date'])
df['time'] = pd.to_datetime(df['time'], format='%H:%M').dt.time
df.set_index('datetime',inplace=True)
df
```

## date selection also time filtered

```python
# df = df.iloc[1:-1]
date = '2022-04-07'
df = df.loc[date] 
df = df.between_time('09:16:00', '15:30:00')
df
```

### calculating range/open_df

```python
"""PART2 - Calculating range"""
#Open range with 15 minutes
start_time = df.index[0]
end_time = df.index[0]+pd.Timedelta(minutes=15)

open_df = df.loc[start_time:end_time]
print(open_df)
```

```python
rangeHigh = open_df['high'].max()
rangeLow = open_df['low'].min()

df['rangeHigh'] = rangeHigh
df['rangeLow'] = rangeLow
```

### range

```python
df['range'] = df['rangeHigh'] - df['rangeLow']
```

## signal function

```python
def get_signal(row):
    if row['high'] > row['rangeHigh']:
        return 'Buy'
    elif row['low'] < row['rangeLow']:
        return 'Sell'
    else:
        return np.nan
    
df['signal'] = df.apply(get_signal,axis=1)
```

```python
df.head(50)
```

## trade position

```python
df['tradePosition'] = df['signal'].fillna(method='ffill')
```

## Trade counter


```python
df['tradeCount'] = 0
current_count = 0

for i in range(len(df)):
    if pd.isnull(df['tradePosition'].iloc[i]):
        df['tradeCount'].iloc[i] = current_count
    else:
        if df['tradePosition'].iloc[i] != df['tradePosition'].iloc[i-1]:
            current_count += 1
        df['tradeCount'].iloc[i] = current_count
```

```python
df.head(50)
```

## Intrade count

```python
df['inTrade'] = np.where((df['tradeCount'] > 0) & (df['tradeCount'] <3), 1, 0)
```

## entryCount

```python
df['entryCount'] = np.where((df['tradeCount']!= df['tradeCount'].shift()) & df['inTrade']==1 , 1,0)
```

## Buy price

```python
df.loc[(df['tradePosition']=='Buy')&(df['inTrade'] ==1),'buyPrice'] = df['rangeHigh']
```

## buySL


```python
condition1 = (df['inTrade'] == 1) & (df['tradePosition'] != 'Buy')
condition2 = (df['inTrade'] == 1) & (df['tradePosition'] == 'Buy')

# df['buySL'] = np.where(condition1, np.NaN, np.where(condition2, df['rangeLow'], df['buyPrice'].shift(1)))
df['buySL'] = np.where(condition1, np.NaN, np.where(condition2, df['rangeLow'], np.NaN))
```

```python
df.head(50)
df.iloc[45:70]
```

## buyTgt

```python
# 20 % not fixed here variable
condition = (df['tradePosition'] == 'Buy') & (df['inTrade'] == 1)
df['buyTgt'] = np.where(condition, df['buyPrice'] + df['range'] * 0.02, np.NaN)
```

## sell price

```python
df.loc[(df['tradePosition']=='Sell')&(df['inTrade'] ==1),'sellPrice'] = df['rangeLow']
```

##   sellSL

```python
condition1 = (df['inTrade'] == 1) & (df['tradePosition'] != 'Sell')
condition2 = (df['inTrade'] == 1) & (df['tradePosition'] == 'Sell')

# df['sellSL'] = np.where(condition1, np.NaN, np.where(condition2, df['rangeHigh'], df['sellPrice'].shift(1)))

df['sellSL'] = np.where(condition1, np.NaN, np.where(condition2, df['rangeHigh'], np.NaN))
```

## sell tgt

```python
condition = (df['tradePosition'] == 'Sell') & (df['inTrade'] == 1)
df['sellTgt'] = np.where(condition, df['sellPrice'] - df['range'] * 0.02, np.NaN)
```

# PART 2 


### buySLHit

```python
df['buySLHit'] = 0
```

```python
df['buySLHit'] = np.where((df['tradePosition'].shift() == 'Buy') & (df['buySLHit'].shift() == 1), 1,
         np.where((df['tradePosition'].shift() == 'Buy') & (df['inTrade'].shift() == 1),
                  np.where(df['low'] <= df['buySL'].shift(), 1, 0),
                  0
                 )
        )
```

## buy_tg hit


```python
df['buyTgtHit'] = 0
```

```python
df['buyTgtHit'] = np.where(
    (df['tradePosition'].shift() == 'Buy') & (df['buyTgtHit'].shift() == 1),
    1,
    np.where(
        (df['tradePosition'].shift() == 'Buy') & (df['inTrade'].shift() == 1) & (df['entryCount'].shift() == 0),
        np.where(df['high'] >= df['buyTgt'].shift(), 1, 0),
        0
    )
)
```

## buyTradeOn

```python
df['buyTradeOn'] = np.where(
    (df['tradePosition'] =='Buy')&(df['inTrade']==1)&(df['buySLHit']==0)&(df['buyTgtHit']==0),
    1,0
    
    )
```

## buyExitPrice

```python
df['buyExitPrice'] = np.where(
        (df['buyTradeOn'].shift() == 1) & (df['buyTradeOn']==0),
        np.where(
                (df['buySLHit']==1), df['buySL'].shift(), df['buyTgt']
        ),0
)
```

## buyMTM

```python
df['buyMTM'] = np.where(df['buyTradeOn' ]==1,df['close'] - df['buyPrice'],0)
```

## Buy PNL 


```python
pd.set_option('display.max_columns',None)
```

```python
df['buyPnL'] = np.where(
    (df['buyTradeOn'].shift() == 1) & (df['buyTradeOn'] == 0),df['buyExitPrice'] - df['buyPrice'].shift(),0)
```

## sellSLHit

```python
df['sellSLHit'] = 0 
```

```python
df['sellSLHit'] = np.where(
    (df['tradePosition'].shift() == 'Sell') & (df['sellSLHit'].shift() == 1),1,
    np.where(
            (df['tradePosition'].shift() == 'Sell') & (df['inTrade'].shift() == 1),
            np.where(df['high'] >= df['sellSL'].shift(),1,0),0
    )
)
```

## sellTgtHit

```python
df['sellTgtHit'] = 0
```

```python
df['sellTgtHit'] = np.where(
    (df['tradePosition'].shift() == 'Sell') & (df['sellTgtHit'].shift() == 1),1,
        np.where(
                    (df['tradePosition'].shift() == 'Sell') & (df['inTrade'].shift() == 1) & (df['entryCount'].shift() == 0),
                    np.where(df['low'] <= df['sellTgt'].shift(),1,0),
                    0
        )
)
```

## sell trade on

```python
df['sellTradeOn'] = np.where(
    (df['tradePosition'] == "Sell") & (df['inTrade'] ==1) & (df['sellSLHit'] == 0) & (df['sellTgtHit'] ==0),1,0
)
```

## 

## sell exit price

```python
df['sellExitPrice'] = np.where(
        (df['sellTradeOn'].shift() ==1)&(df['sellTradeOn'] == 0),
            np.where(df['sellSLHit'] ==1,df['sellSL'],df['sellTgt'])
        ,0
)
```

## SEllMTM

```python
df['sellMTM'] = np.where(df['sellTradeOn'] ==1,df['sellPrice'] - df['close'],0)
```

## sellPnL

```python
df['sellPnL'] = np.where(
    (df['sellTradeOn'].shift() == 1) & (df['sellTradeOn'] == 0),
    df['sellPrice'].shift() - df['sellExitPrice'],
    0
)
```

# Part 3

### buy max mtm

```python
         
df['buyMaxMTM'] = np.where(
    (df['tradePosition'] =="Buy") & (df['inTrade']==1) & (df['high'] - df['buyPrice'] > 0),(df['high'] - df['buyPrice']),0
)
```

## buy min mtm

```python
df['buyMinMTM'] = np.where(
    (df['tradePosition'] =="Buy") & (df['inTrade']==1) & (df['high'] - df['buyPrice'] < 0),(df['high'] - df['buyPrice']),0
)
```

## Sell max mtm

```python
df['sellMaxMTM'] = np.where(
    (df['tradePosition'] =="Sell") & (df['inTrade']==1) & (df['sellPrice'] - df['low'] > 0),(df['sellPrice'] - df['low']),0
)
```

##  sell min mtm

```python
df['sellMinMTM'] = np.where(
    (df['tradePosition'] =="Sell") & (df['inTrade']==1) & (df['sellPrice'] - df['low'] < 0),(df['sellPrice'] - df['low']),0
)
```

##  reading files

```python
import os
import pandas as pd

folder_name = "C:\\Users\\Admin\\OneDrive\\Desktop\\Nifty50_data\\Nifty50_Data"
dfs = []

# loop through each file in the folder
for filename in os.listdir(folder_name)[0:2]:
    if filename.endswith('.csv'):
        # get the full file path of the file to read
        file_path = os.path.join(folder_name, filename)

        # read the file data into a DataFrame
        df = pd.read_csv(file_path)

        # store the DataFrame in the list
        dfs.append(df)

# print the resulting DataFrames
for df in dfs:
    print(df)
```

## stg-2

```python
open_day = 4
closing_day = 31

a = dfs[0].apply(lambda row: 1 if row['day'] >= open_day else 0, axis=1)
b = dfs[0].apply(lambda row: 1 if row['day'] <= closing_day else 0, axis=1)

mask = (a == 1) & (b == 1)
x = dfs[0][mask]
```

```python
x.head(35)
```

```python
x = x.resample('M').agg({'Open':'first','High':'max','Low':'min','Adj Close':'last'})
```

```python
x.head(40)
```

```python
x['Monthly return'] = (x['Adj Close']/x['Open'] -1)*100
```

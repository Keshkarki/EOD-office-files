```python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime as dt
```

## ticker list

```python
series = pd.read_csv('symbol.csv', header=None).squeeze()
tickers = series.apply(lambda ticker : f"{ticker}.NS").to_list()
print(tickers)
```

##  companies list

```python
#for making columns
company_list = series.apply(lambda ticker : f"{ticker.split('.')[0]}").to_list()
print(company_list)
```

## downloading all nifty stocks 'Adj Close' price

```python
start_date = '2008-01-01'
end_date = '2023-06-22'

#data holder dataframe
data = pd.DataFrame(columns=company_list)

for ticker in tickers:
    data[ticker.split('.')[0]] = yf.download(ticker,start_date,end_date)['Adj Close']
```

## downloading nifty file


```python
nifty = yf.download('^NSEI',start_date, end_date)
nifty
```

```python
data.insert(0,'Nifty',nifty['Adj Close'])
```

```python
data = data.round(2)
```
